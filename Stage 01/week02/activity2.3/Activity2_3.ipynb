```python
import matplotlib.pyplot as plt
import numpy as np

# Define x range and y = sin(x)
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# Create the plot
plt.figure(figsize=(8, 4))
plt.plot(x, y, label='y = sin(x)', color='blue', linestyle='--', linewidth=2)

# Add labels, title, grid, and legend
plt.title('Plot of y = sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
