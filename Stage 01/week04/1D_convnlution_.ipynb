```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

Load the image using OpenCV

```python
!pwd
```

```python
img = cv2.imread('/content/sample_data/emma.jpg')
img.shape
```

Convert the RGB image into a grayscale image


```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img.shape
```

Define vertical and horizontal filter kernels

```python
v_kernel = np.array([
  [-1, 0, 1],
  [-1, 0, 1],
  [-1, 0, 1]
])
```

Do the filtering

```python
v_image = cv2.filter2D(img, -1, v_kernel)
```

Display results


```python
plt.imshow(img, cmap="gray")
plt.title('Orginal Image')

plt.figure()
plt.imshow(v_image, cmap="gray")
plt.title('Vertical Image')
```

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

im = cv2.imread('/content/sample_data/emma.jpg')
im.shape

im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
im.shape

h_kernel = np.array([
  [-1, -1, -1],
  [0, 0, 0],
  [1, 1, 1]
])

h_image = cv2.filter2D(im, -1, h_kernel)

plt.imshow(im, cmap="gray")
plt.title('Orginal Image')

plt.figure()
plt.imshow(h_image, cmap="gray")
plt.title('Horizontal Image')
```
